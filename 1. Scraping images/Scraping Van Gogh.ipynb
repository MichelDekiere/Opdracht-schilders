# Scraping Van Gogh schilderijen

Net zoals bij Rembrandt, zal ik van Van gogh ook wat schilderijen scrapen. Ik zal in deze notebook zoveel mogelijk images scrapen van de schilder Vincent Van Gogh. Ik zal de scraping uitvoeren op [deze](http://vggallery.com/) site.

```python
import requests
from bs4 import BeautifulSoup
import os
import urllib
```

Nieuwe folder maken om de gescrapte images in op te slaan, we maken deze dan ook de actieve folder zodat de images dan ook daadwerkelijk hierin opgeslagen worden.

```python
os.mkdir("../datasets/paintings/Van Gogh")  # aanmaken directory
os.chdir("../datasets/paintings/Van Gogh/")  # deze directory active maken, zodat de gescrapete images hierin terecht komen
```

```python
base_url = "http://vggallery.com/painting/"
pages = ["main_ac.htm", "main_df.htm", "main_gi.htm", "main_jl.htm", "main_mo.htm", "main_pr.htm", "main_su.htm", "main_vz.htm"]

count = 1

for page in pages:
  response = requests.get(base_url+ page)
  soup = BeautifulSoup(response.text, 'html.parser')

  # Iedere link ophalen op de pagina
  links = soup.find_all('a')

  # Over iedere link itereren
  for link in links:
    try:
        href = link.get('href')
        if href.startswith('p_'): # wanneer de webpagina begint met 'p_' in de naam, is het een schilderij pagina
            image_page_link = base_url + href # volledige link van de pagina opmaken
            response = requests.get(image_page_link)
            soup = BeautifulSoup(response.text, "html.parser")

            # img tag ophalen
            image = soup.find_all("img")[2] # iedere keer 3e image op de pagina
            image_src = image["src"] # src attribuut van de image ophalen

            image_link = base_url + image_src # volledige link van de image op de site opstellen

            try:
                file_name = f"Van Gogh {count}.jpg" # formaat van image naam --> Van Gogh 1.jpg, Van Gogh 2.jpg, ...
                urllib.request.urlretrieve(image_link, file_name) # downloaden van de image
                print(image_link) # printen van succesvolle images
                count += 1
            except:
                pass
    except:
      pass
```

```python
print(count)
```
